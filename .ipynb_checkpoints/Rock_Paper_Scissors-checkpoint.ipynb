# Collect the Data

#### Function that launches camera to collect data

```python
def collect_data(num_samples):
    global rock, paper, scissor, nothing
    
    # Initialise the camera
    cap = cv2.VideoCapture(0,  cv2.CAP_DSHOW)
    cap.set(3, 640)
    cap.set(4, 480)

    # Initialise to false
    trigger = False
    
    # Counter for number of samples collected
    counter = 0
    
    # ROI size
    box_size = 234
    
    # Get the width of the frame from default camera properties
    width = int(cap.get(3))


    while True:
        # Read frame by frame
        ret, frame = cap.read()
        
        # Flip the frame horizontally
        frame = cv2.flip(frame, 1)
        
        # Break the loop if the frame isn't read properly
        if not ret:
            break
            
        # If counter is equal to the number of samples...
        if counter == num_samples:
            # Reset the trigger and counter
            trigger = not trigger
            counter = 0
        
        # Define ROI for capturing samples
        cv2.rectangle(frame, (width - box_size, 0), (width, box_size), (38, 72, 249), 2)
        
        # If trigger = true, begin sample collection
        if trigger:
            
            # Collect data only from the selected ROI
            roi = frame[5: box_size-5 , width-box_size + 5: width -5]
            
            # Append "roi" and "class_name" to the list
            eval(class_name).append([roi, class_name])
            
            counter += 1 
            text = "Collected Samples of {}: {}".format(class_name, counter)
            
        else:
            text = "Please press one of the listed keys."
        
        # Show the counter on the frame
        cv2.putText(frame, text, (180, 460), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 1, cv2.LINE_AA)
        
        # Read the background image
        samplingBG = cv2.imread("images/SamplingScreen.png")
        
        # Scale the frame
        frameScaled = cv2.resize(frame, (0, 0), None, 0.63, 0.63)
        frameScaled = frameScaled[:,101:403]
        
        # Put frame on the background image
        samplingBG[213:515,473:775] = frameScaled
        
        # Display the windows
        cv2.imshow("Play: Rock Paper Scissors!", samplingBG)
        
        k = cv2.waitKey(1)
        
        # If the user presses one of the following keys, set class_name to respective label, set trigger to true
        # R: class_name = rock
        if k == ord('r') or k == ord('R'):
            trigger = not trigger
            class_name = 'rock'
            rock = []
            
        # P: class_name = paper  
        if k == ord('p') or k == ord('P'):
            trigger = not trigger
            class_name = 'paper'
            paper = []
        
        # S: class_name = scissor  
        if k == ord('s') or k == ord('S'):
            trigger = not trigger
            class_name = 'scissor'
            scissor = []
                    
        # N: class_name = rock
        if k == ord('n') or k == ord('N'):
            trigger = not trigger
            class_name = 'nothing'
            nothing = []
        
        # Exit if user presses 'Q'
        if k == ord('q') or k == ord('Q'):
            break
            
    cap.release()
    cv2.destroyAllWindows()

# Run the function
no_of_samples = 100
collect_data(no_of_samples)
```

# Model Training

#### Preprocessing data

```python
# Combine the labels of all classes together
labels = [x[1] for x in rock] + [x[1] for x in paper] + [x[1] for x in scissor] +[x[1] for x in nothing]

# Combine the images of all classes together
images = [x[0] for x in rock] + [x[0] for x in paper] + [x[0] for x in scissor] +[x[0] for x in nothing]
```

```python
# Print out the total number of labels and images.
print('Total images: {} , Total Labels: {}'.format(len(labels), len(images)))
```

```python
# Normalize the images by scaling to a range between 0 and 1
images = np.array(images, dtype="float") / 255

# Create an encoder Object
encoder = LabelEncoder()

# Convert Labels to integers in alphabetical order
# nothing= 0, paper = 1, rock = 2, scissor = 3
Int_labels = encoder.fit_transform(labels)

# Convert int_labels to one-hot
one_hot_labels = to_categorical(Int_labels, 4)
```

```python
# Splitting data into 75% and 25% ratio: train = 75% and test = 25%
(X_train, X_test, y_train, y_test) = train_test_split(images, one_hot_labels, test_size=0.25, random_state=50)

# Empty memory from RAM
images = []
```

#### Creating and fine tuning the model

```python
# the input size matches the size of the captured ROI
image_size = 224

# Create an instance of NASNetMobile model without the head
N_mobile = tf.keras.applications.NASNetMobile( input_shape=(image_size, image_size, 3), include_top=False, weights='imagenet')

# Freeze all layers of the NASNetMobile model
N_mobile.trainable = False
    
# Connect the output of the NASNetMobile model to a GlobalAveragePooling2D layer
x = N_mobile.output
x = GlobalAveragePooling2D()(x)

# Add a Dense layer with 712 neurons and ReLU activation
x = Dense(712, activation='relu')(x) 

# a Dropout layer with a rate of 0.40 to reduce overfitting
x = Dropout(0.40)(x)

# Connect the output of the Dropout layer to a Dense layer with 4 neurons and 'softmax' activation
preds = Dense(4,activation='softmax')(x) 

# Create the final model by connecting the input of NASNetMobile model to the output of the Dense layer
model = Model(inputs=N_mobile.input, outputs=preds)

# Print the number of layers in the model
print ("Number of Layers in Model: {}".format(len(model.layers[:])))
```

#### Training the Model

```python
# Transformations to ensure that the model works in as many scenarios as possible
augment = ImageDataGenerator(
        channel_shift_range=0.10, # randomly shift color channels
        width_shift_range=0.10, # randomly shift images horizontally (fraction of total width)
        height_shift_range=0.10, # randomly shift images vertically (fraction of total height)
        rotation_range=30, # randomly rotate images in the range (degrees, 0 to 180)
        zoom_range=0.25, # Randomly zoom image 
        shear_range=0.10, # shear intensity (shear angle in counter-clockwise direction as radians)
        horizontal_flip=True, # randomly flip images
        fill_mode="nearest" # method used for filling in newly created pixels
)

# Cannot add bridhtness_range as it would lead to overfitting
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Set batchsize according to your system
epochs = 10
batchsize = 20

# Start training
history = model.fit(x=augment.flow(X_train, y_train, batch_size=batchsize), validation_data=(X_test, y_test), 
steps_per_epoch= len(X_train) // batchsize, epochs=epochs)
```

# Model Evaluation

```python
# Importing all necessary libraries
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
# Show accuracy without formatting
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test accuracy: {:.2f}%".format(test_acc * 100))
```

```python
# Store the accuracy values of the training and validation set
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

# Epochs
epochs = range(len(acc))
```

```python
# Create a dataframe with the accuracy and loss values
data = {'epoch': epochs, 'acc': acc, 'val_acc': val_acc, 'loss': loss, 'val_loss': val_loss}
df = pd.DataFrame(data)

# Create a heatmap of the accuracy values
plt.figure(figsize=(10,4))
sns.heatmap(df[['acc', 'val_acc']], annot=True, cmap='YlGnBu')
plt.title('Accuracy heatmap')
plt.xlabel('Accuracy')
plt.ylabel('Epoch')

# Create a heatmap of the loss values
plt.figure(figsize=(10,4))
sns.heatmap(df[['loss', 'val_loss']], annot=True, cmap='YlGnBu')
plt.title('Loss heatmap')
plt.xlabel('Loss')
plt.ylabel('Epoch')

plt.show()
```

```python
# Plot the accuracy values 
plt.plot(epochs, acc, 'b', label='acc')
plt.plot(epochs, val_acc, 'r', label='val_acc')
plt.title('Accuracy')
plt.legend()

# Create a new figure
plt.figure()

# Plot the loss values
plt.plot(epochs, loss, 'b', label='loss')
plt.plot(epochs, val_loss, 'r', label='val_loss')
plt.title('Loss')
plt.legend()

plt.show()
```

# Playable Game

#### Function to identify the winner

```python
# Basic rock paper scissors logic
def findout_winner(user_move, Computer_move):
    
    # If user and computer have the same move, it's a tie
    if user_move == Computer_move:
        return "Tie"
    
    # if user plays rock and computer plays scissor, user wins
    elif user_move == "rock" and Computer_move == "scissor":
        return "User"
    
    # if user plays rock and computer plays paper, computer wins
    elif user_move == "rock" and Computer_move == "paper":
        return "Computer"
    
    # if user plays scissor and computer plays rock, computer wins
    elif user_move == "scissor" and Computer_move == "rock":
        return "Computer"
    
    # if user plays scissor and computer plays paper, user wins
    elif user_move == "scissor" and Computer_move == "paper":
        return "User"
    
    # if user plays paper andc computer plays rock, user wins
    elif user_move == "paper" and Computer_move == "rock":
        return "User"
    
    # if user plays paper and computer plays scissors, computer wins
    elif user_move == "paper" and Computer_move == "scissor":
        return "Computer"
```

#### Function to display the game results

```python
def show_results(user_score, computer_score):    
    
    # if the user has a higher score, show the winning screen
    if user_score > computer_score:
        img = cv2.imread("images/YouWin.png")
    
    # if the computer has a higher score, show the losing screen
    elif user_score < computer_score:
        img = cv2.imread("images/YouLose.png")
    
    # otherwise, if they're equal, show the draw screen
    else:
        img = cv2.imread("images/Draw.png")
    
    # display "img" on the "Play: Rock Paper Scissors!" window
    cv2.imshow("Play: Rock Paper Scissors!", img)
    
    # If the user presses 'ENTER' key then return TRUE, else, FALSE
    k = cv2.waitKey(0)
    if k == 13:
        return True 

    else:
        return False
```

#### Function to display the computer's move

```python
def display_computer_move(computer_move_name, uiBG):
    
    # read the corresponding image according to the name of the move
    icon = cv2.imread( "images/{}.png".format(computer_move_name), 1)
    icon = cv2.resize(icon, (302,302))
    
    # display it on the uiBG with these coordinates
    uiBG[213:515, 106:408] = icon

    return uiBG
```

#### Final Application

```python
cap = cv2.VideoCapture(0)
box_size = 234
width = int(cap.get(3))

# Read the background image
uiBG = cv2.imread("images/UIScreen.png")

# List of labels for the classes being predicted
label_names = ['nothing', 'paper', 'rock', 'scissor']


# Number of rounds in the game
attempts = 5

# Initialise moves as "nothing"
computer_move_name= "nothing"
final_user_move = "nothing"

# Initialise all scores as 0
computer_score, user_score = 0, 0

# At each iteration we will decrease the total_attempts value by 1
total_attempts = attempts


# ROI border color
rect_color = (38, 72, 249)

# Initialise this as false
hand_inside = False

# Ignore predictions with probabilities lower than 70%
confidence_threshold = 0.70

# Takes the average of 5 predictons by using a deque object to smoothen the output
smooth_factor = 5

# Initial deque list will have 'nothing' repeated 5 times.
de = deque(['nothing'] * 5, maxlen=smooth_factor)

while True:
    # Read frame by frame
    ret, frame = cap.read()
    
    # Flip the frame horizontally
    frame = cv2.flip(frame, 1)
    
    # Break the loop if the frame isn't read properly
    if not ret:
        break

    # Collect data only from the selected ROI
    roi = frame[5: box_size-5 , width-box_size + 5: width -5]
    
     # Normalize the image and convert to float64 array
    roi = np.array([roi]).astype('float64') / 255.0
    
    # Predict the move
    pred = model.predict(roi)
    
    # Index of the predicted class
    move_code = np.argmax(pred[0])
   
    # Class name of the predicted class
    user_move = label_names[move_code]
    
    # Confidence of the predicted class
    prob = np.max(pred[0])
    
    # Check if probability is above 70%
    if prob >= confidence_threshold:
        
        # Now add the move to deque list from left
        de.appendleft(user_move)
        
        # Get the mode from the last 5 moves
        try:
            final_user_move = st.mode(de)[0][0] 
            
        except StatisticsError:
            print('Stats error')
            continue
             
        # If nothing is untrue and there is a hand inside the box, proceed
        if final_user_move != "nothing" and hand_inside == False:
            
            # Set hand inside to True
            hand_inside = True 
            
            # Computer's Move
            computer_move_name = choice(['rock', 'paper', 'scissor'])
            
            # User's Move
            winner = findout_winner(final_user_move, computer_move_name)
            
            # Display the computer's move
            display_computer_move(computer_move_name, uiBG)
            
            # Decrease attempts by 1
            total_attempts -= 1
            
            # If computer wins, increase computer score and change ROI border to red
            if winner == "Computer":
                computer_score +=1
                rect_color = (0, 0, 255)
            
            # If user wins, increase user score and change ROI border to green
            elif winner == "User":
                user_score += 1;
                rect_color = (0, 250, 0)
                
            # If tie, change ROI border to white
            elif winner == "Tie":
                rect_color = (255, 250, 255)
                
                
            # If user runs out of attempts, proceed     
            if total_attempts == 0:
                
                play_again = show_results(user_score, computer_score)
                
                # If user plays again, reinitialise the variables, else quit
                if play_again:
                    user_score, computer_score, total_attempts = 0, 0, attempts
                
                else:
                    break
        
        # Display computer move when a hand is inside the box
        elif final_user_move != "nothing" and hand_inside == True:
            display_computer_move(computer_move_name, uiBG)
    
        # If the box captures nothing, then hand_inside becomes false
        elif final_user_move == 'nothing':            
            hand_inside = False
            rect_color = (38, 72, 249)
 

    # Display moves, scores, and attempts on the frame
    cv2.putText(frame, "Your Move: " + final_user_move,
                    (180, 420), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (38, 72, 249), 1, cv2.LINE_AA)

    cv2.putText(frame, "Computer's Move: " + computer_move_name,
                (180, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (38, 72, 249), 1, cv2.LINE_AA)

    cv2.putText(frame, "Your Score: " + str(user_score),
                    (180, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (38, 72, 249), 1, cv2.LINE_AA)
    cv2.putText(frame, "Computer Score: " + str(computer_score),
                    (180, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (38, 72, 249), 1, cv2.LINE_AA)
    
    cv2.putText(frame, "Attempts left: {}".format(total_attempts), (430, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.7,
                (190, 160, 30), 1, cv2.LINE_AA)
    
    # Define ROI for capturing gestures
    cv2.rectangle(frame, (width - box_size, 0), (width, box_size), rect_color, 2)
        
    # Scale the frame
    frameScaled = cv2.resize(frame, (0, 0), None, 0.63, 0.63)
    frameScaled = frameScaled[:,101:403]
    
    # Put frame on the background image
    uiBG[213:515,473:775] = frameScaled

    # Display the windows
    cv2.imshow("Play: Rock Paper Scissors!", uiBG)

    # Exit if 'q' is pressed 
    k = cv2.waitKey(10)
    if k == ord('q') or k == ord('Q'):
        break

cap.release()
cv2.destroyAllWindows()
```
